```python
import librosa
import numpy as np
import os
import matplotlib.pyplot as plt
import soundfile as sf

# Definieren der Variablen
source_dir = "data/noises/noise_2.wav"
source_dir_2 = "data/knocks/knock_3_var1.wav"
y, sr = librosa.load(source_dir)
y_2, sr_2 = librosa.load(source_dir_2)

# Berechnen von MFCCs und Mel-Spektrogramm
mfcc = librosa.feature.mfcc(y=y, sr=sr)
mfcc_2 = librosa.feature.mfcc(y=y_2, sr=sr_2)
spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
spectrogram_2 = librosa.feature.melspectrogram(y=y_2, sr=sr_2)
print("Spectrogram:", spectrogram.shape)
print("Spectrogram_2:", spectrogram_2.shape)
print("MFCCs:", mfcc.shape)
print("MFCC_2s:", mfcc_2.shape)
print(len(y), len(y_2))


# Plot the spectrogram
plt.figure(figsize=(10, 4))
plt.imshow(spectrogram, cmap='hot', origin='lower', aspect='auto')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()


# Plot the MFCC data
plt.figure(figsize=(10, 4))
plt.imshow(mfcc, cmap='hot', origin='lower', aspect='auto')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.tight_layout()
plt.show()

# Plot the MFCC data
plt.figure(figsize=(10, 4))
plt.imshow(mfcc_2, cmap='hot', origin='lower', aspect='auto')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.tight_layout()
plt.show()

overlapped_y = y + y_2
# sf.write("output.wav", overlapped_y, sr)


```
