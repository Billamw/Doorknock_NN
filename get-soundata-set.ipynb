```python
import soundata
from torch.utils.data import DataLoader
```

```python
# Initialize the FSD50K dataset
dataset = soundata.initialize('fsd50k', data_home='data/fsd50k')
dataset.download()
```

```python
dataset_dict = dataset.load_clips()
```

```python
dataset_dict[next(iter(dataset_dict))].audio
```

```python
dataset.clip('47608').tags
```

```python
# Angenommen, dataset_dict ist Ihr ursprüngliches Dictionary
# und jedes Element in dataset_dict hat eine 'tags' Liste

knocks_dict = {}

for key, value in dataset_dict.items():
    print(value.tags.labels)
    if 'Knock' in value.tags.labels and 'Door' in value.tags.labels:
        knocks_dict[key] = value

# Jetzt enthält knocks_dict alle Elemente, deren 'tags' "knock" enthalten
```

```python
knocks_dict
```

```python
import csv
import os
import shutil

# Erstellen des neuen Ordners für WAV-Dateien, falls er nicht existiert
new_knocks_folder = 'data/new_knocks'
os.makedirs(new_knocks_folder, exist_ok=True)

# Pfad zur CSV-Datei
csv_file_path = 'data/knocks_data.csv'

# Öffnen der CSV-Datei zum Schreiben
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    # Schreiben der Kopfzeile der CSV-Datei
    csv_writer.writerow(['Clip ID', 'Audio Path', 'Tags'])

    # Durchgehen jedes Eintrags in knocks_dict
    for clip_id, clip_data in knocks_dict.items():
        # Pfad zur aktuellen WAV-Datei
        audio_path = clip_data.audio_path
        # Tags des Clips
        tags = clip_data.tags.labels
        
        # Schreiben der Clip-Daten in die CSV-Datei
        csv_writer.writerow([clip_id, 'data/new_knocks/' + clip_id + '.wav', ';'.join(tags)])
        
        # Kopieren der WAV-Datei in den neuen Ordner
        shutil.copy(audio_path, os.path.join(new_knocks_folder, os.path.basename(audio_path)))

print("Daten erfolgreich in CSV gespeichert und WAV-Dateien kopiert.")
```
